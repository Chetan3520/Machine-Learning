# Step4

# training the various Algorithms

## Splitting Data into Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,random_state=0)
```

## Scalling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

## Validating The Models

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

## Validation Libraries

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
```

```python
models=[]
models.append(("LDA",LinearDiscriminantAnalysis()))
models.append(("LR",LogisticRegression(solver="liblinear",multi_class="ovr")))
models.append(("DT",DecisionTreeClassifier()))
models.append(("KNN",KNeighborsClassifier()))
models.append(("CART",DecisionTreeClassifier()))
models.append(("SVM",SVC()))
models.append(("RandF",RandomForestClassifier()))
```

```python
name=[]
res=[]
names=[]
for name,model in models:
    skfold=StratifiedKFold(n_splits=10,random_state=None)
    cv_result=cross_val_score(model,X_train,y_train,cv=skfold,scoring='accuracy')
    res.append(cv_result.mean())
    names.append(name)
    print('%s: %f: (%f)'%(name,cv_result.mean(),cv_result.std()))
plt.bar(names,res,color="pink",width=0.6)
plt.title("Algorithm Comparison")
plt.show()
    
    
```

# Conclusion:

So we can observe and see the result and performance of the above classification models 
1. Logistic Regression Classifier have maximum accuracy with some possible cross validations.
2. If we perform parameter tunning for SVM them we can also get the more accurate model.


# Step5

# Train the best fit Model

```python
l=['newton-cg', 'sag', 'saga' ,'lbfgs']
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,y_train)
skfold=StratifiedKFold(n_splits=10,random_state=None)
cv_result=cross_val_score(model,X_train,y_train,cv=skfold,scoring='accuracy')
print(cv_result.mean()*100,'%')
```

# Step6

# Making Predictions`

```python
y_pred=model.predict(X_test)
```

# Step7

# Validation of The Best Fit Model

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print("ACCURACY SCORE:%f  \n"%accuracy_score(y_test,y_pred))
print('CONFUSION MATRIX: \n',confusion_matrix(y_test,y_pred))
print('\nCLASSIFICATION REPORT:\n ',classification_report(y_test,y_pred))
```
