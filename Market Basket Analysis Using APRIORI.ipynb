# Data Preprocessing

```python
m,n=df.shape
```

```python
#we just want to take the data which is non empty so we will combine whole data as list
transaction=[]
for i in range(m):
    transaction.append([str(df.values[i,j]) for j in range(n)])
transaction
```

# Trainnig The APRIORI Algorithm

# Measures:


1. __Support__: Signifies the Popularity of the item 
    

Sup(item)=(No. of Transaction of items)/(Total no. of Transactions)

2. __Confidence__:Signifies the likelihood of the item1 purchased when item2 purchased

conf(item1-->item2)=Sup(item1 U item2)/sup(item1)

3. __lift__: there may be equal probabilities of item1 and item2 which increase the confidence, to overcome this we use lift

lift(item1-->item2=Sup(item1 U item2)/(sup(item1)*sup(item2))

```python
#!pip install apyori
```

```python
from apyori import apriori
rules=apriori(transactions=transaction,min_support=0.003,min_confidence=0.2,min_lift=3,min_length=2,max_length=2)
```

# Result

```python
results=list(rules)
results
```

# Results in Dataframe

```python
lhs         = [tuple(result[2][0][0])[0] for result in results]
rhs         = [tuple(result[2][0][1])[0] for result in results]
supports    = [result[1] for result in results]
confidences = [result[2][0][2] for result in results]
lifts       = [result[2][0][3] for result in results]
resultsinDataFrame = pd.DataFrame(zip(lhs, rhs, supports, confidences, lifts), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
resultsinDataFrame
```
