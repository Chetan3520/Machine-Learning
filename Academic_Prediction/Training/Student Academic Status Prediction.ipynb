```python
from scipy import stats as ss
from itertools import product
import pandas as pd
categorical = ['marital_status', 'application_mode', 'course',
       'mor_eve_attendance', 'prev_qualification', 'prev_qualification_grade',
       'nationality', 'mother_qualification', 'father_qualification',
       'mother_occupation', 'father_occupation', 'admission_grade',
       'displaced', 'edu_special_needs',
       'gender','target']

l1=categorical
l2=categorical
cat_var_product=list(product(l1,l2,repeat=1))

result=[]
for i in cat_var_product:
    if i[0]!=i[1]:
        result.append((i[0],i[1],list(ss.chi2_contingency(pd.crosstab(df2[i[0]],df2[i[1]])))[1]))

chi_sq_output=pd.DataFrame(result,columns=['l1','l2','p_value'])
chi_sq_output.head()

print('Given tables shows the pvalue corresponding two categorical variables')
chi_sq_output.pivot(index='l1',columns='l2',values='p_value')
```

__Conclusion:__
